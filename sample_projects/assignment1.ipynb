```python
import pandas as pd
import csv
```

```python
file = open("F:/sumeet_prasad/text_input.txt", 'rb')
question = []
answer = []
for line in file:
    line = line.lower().strip()
    line = line.decode("ascii","ignore")
    if line.endswith("?"):
        question.append(line)
    elif line.startswith("answer"):
        answer.append(line)
file.close()
```

```python
answer
```

```python
df = pd.DataFrame(data=list(zip(question,answer)),columns=['Question','Answer'])
```
