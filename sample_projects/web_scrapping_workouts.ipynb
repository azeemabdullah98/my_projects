```python
import requests
from bs4 import BeautifulSoup
import re
import os
import pandas as pd
```

```python
start_url="https://www.fipola.in/"
```

```python
source=requests.get(start_url).text
```

```python
source
```

```python
soup=BeautifulSoup(source,'lxml')
```

```python
soup
```

```python
product_links=[]
for links in soup.find_all("div" ,class_="fivecols"):
    product_links.append(links.find("a")['href'])
```

```python
for links in product_links:
    Response=requests.get(links).text
    html=BeautifulSoup(Response,'lxml')
    nameList=[]
    for title in html.find_all("div",itemprop="description"):
        titles=re.sub("[\W]","",title.a.text)
        nameList.append(titles)
    weightList=[]
    for weight in html.find_all("div",itemprop="description"):
        weight=re.sub("[^a-zA-Z0-9-]","",weight.ul.text)
        weightList.append(weight)
    priceList=[]
    for price in html.find_all("div",itemprop="description"):
        prices=price.find("span",class_="price").text
        priceList.append(prices)
    df=pd.DataFrame(list(zip(nameList,weightList,priceList)),columns=['Name','Weight','Price'])
    path="c:/python_projects/tendercuts/fipola.csv"
    if os.path.isfile(path) == True:
        file=pd.read_csv(path)
        data=pd.concat([file,df],ignore_index=True)
        data.to_csv(path,index=False)
    else:
        print("creating new file...")
        df.to_csv(path,index=False)
    
```
